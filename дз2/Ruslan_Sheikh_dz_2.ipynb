**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
def loop_11_1():
  sum_writing_score = {}
  count = {}
  for i in range(len(df)):
    sum_writing_score[df["race/ethnicity"].iloc[i]] = sum_writing_score.get(df["race/ethnicity"].iloc[i],0)+df["reading score"].iloc[i]
    count[f"{df['race/ethnicity'].iloc[i]} len"] = count.get(f"{df['race/ethnicity'].iloc[i]} len",0)+1
  res={}
  for i in sum_writing_score:
    res[i] = sum_writing_score[i]/count[f"{i} len"]
  return res

res = loop_11_1()
for i in res:
  print(i,res[i])
%timeit res = loop_11_1()
```

```python
def loop_11_2():
  min_scores = {}
  for i in range(len(df)):
    min_scores[df["education"].iloc[i]] = min(min_scores.get(df["education"].iloc[i],10e10), df["writing score"].iloc[i])
  res = {}
  for i in min_scores:
    res[i] = min_scores[i]
  return res

res = loop_11_2()
for i in res:
  print("%-20s" % i,res[i])
%timeit res = loop_11_2()
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
df["Math_PassStatus"] = df["math score"].map(lambda x: "F"*(x<passmark) + "P"*(x>=passmark))
df['Math_PassStatus'].value_counts()
```

```python
df["Reading_PassStatus"] = df["reading score"].map(lambda x: "F"*(x<passmark) + "P"*(x>=passmark))
df['Reading_PassStatus'].value_counts()
```

```python
df["Writing_PassStatus"] = df["writing score"].map(lambda x: "F"*(x<passmark) + "P"*(x>=passmark))
df['Writing_PassStatus'].value_counts()
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
def good_scores (x):
  res = (x["math score"]>=passmark) & (x["reading score"]>=passmark) & (x["writing score"]>=passmark)
  return res*"P"+ (not res)*"F"


df["OverAll_PassStatus"] = df.apply(good_scores, axis=1)
df["OverAll_PassStatus"].value_counts()
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
grade = {50:'F', 60:'E', 70:'D', 80:'C', 90:'B'}

def GetGrade(average_mark):
  for i in grade:
    if average_mark<i:
      return grade[i]
  return 'A'

df["Grade"] = df.apply(lambda x: GetGrade((x["math score"]+x["reading score"]+x["writing score"])/3), axis=1)
df["Grade"].value_counts()
```
